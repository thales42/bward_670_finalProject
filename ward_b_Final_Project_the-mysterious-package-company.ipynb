## Content Analytics for - The Mysterious Package Company (topic modeling)

```python
from gensim.corpora import Dictionary
from gensim.models import ldamodel

from gensim.models.coherencemodel import CoherenceModel
from gensim.models.wrappers import LdaVowpalWabbit, LdaMallet

import warnings
warnings.filterwarnings('ignore')  # To ignore all warnings that arise here to enhance clarity
```

```python
df2 = df1[["comment_message", "comment_likes"]].copy()
df2.head()
```

```python
texts = pd.Series(df2["comment_message"])
```

```python
texts[:30]
```

```python
texts = texts.tolist()
#texts
```

```python
# Remove useless numbers and alphanumerical words
documents = [re.sub("[^a-zA-Z]+", " ", text) for text in texts]
# tokenize
texts = [[word for word in text.lower().split() ] for text in documents]
# stemming words: having --> have; friends --> friend
lmtzr = WordNetLemmatizer()
texts = [[lmtzr.lemmatize(word) for word in text ] for text in texts]
# remove common words 
stoplist = stopwords.words('english')
texts = [[word for word in text if word not in stoplist] for text in texts]
#remove short words
texts = [[ word for word in tokens if len(word) >= 3 ] for tokens in texts]
```

```python
# this is text processing required for topic modeling with Gensim
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.wrappers import LdaVowpalWabbit, LdaMallet

for k in range(2,10):
    goodLdaModel = ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=k, passes=50)
    goodcm = CoherenceModel(model=goodLdaModel, corpus=corpus, dictionary=dictionary, coherence='u_mass')
    print goodcm.get_coherence()
```

```python
numpy.random.seed(1) 
k_range = range(5,20,5)   
scores = []
for k in k_range:
    print k         
    goodLdaModel = ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=k, passes=75)  #for better results, use a higher passes (e.g., 50)
    goodcm = CoherenceModel(model=goodLdaModel, corpus=corpus, dictionary=dictionary, coherence='u_mass')
    scores.append(goodcm.get_coherence())
    
plt.figure(figsize=(14, 8))
plt.plot(k_range, scores)
```

## Conclusions:

**The Mysterious Package Company** uses Facebook, Twitter, and Instagram for three distinct different uses in marketing their product through social media. *Facebook* has the largest number of followers (367427). Remarkably, The Mysterious Package Company only has 443 public posts with a stunning 87822 *likes*. This gives them a *like* to *post* ratio of 54%.

When looking at word frequencies and topic modeling of the posts, it becomes evident that Facebook is almost exclusively used for product release announcements. In fact, there are very few references that indicate themes or storylines of products. Principally, Facebook is used to inform their audience to look at The Mysterious Package Company website for new releases.

*Twitter*, on the other hand, is used almost exclusively for replies to customers. Of the 792 tweets, there are 584 replies to other Twitter users. Following any of the links in the sampled replies leads almost always to some form of customer support. Word frequency bi-grams bears this out. Most of the other types of analysis on The Mysterious Package Company tweets provided little information.

But, The Mysterious Package Company's use of *Instagram* is where we find the masterful storytelling that has been the company’s trademark. Although we find only a fraction of the followers that Facebook has (9121 followers), every post is a carefully structured piece of hint fiction. Cryptic images and subtle word usage lure the audience to dig deeper and consider subscribing to another package release. Oddly, the sentiment analysis of followers' comments tended to be neutral with leanings to the positive (very few negative comments...). This may be attributed to fans adopting the indirect language that is the hallmark of hint fiction. Clustering, however, did reveal a number of potential narratives (optimal k-value of 18 clusters). This is particularly intriguing because The Mysterious Package Company reveals very little of their product lineup. Without membership, it is almost impossible to find more than the most recent packages available. If we are to take the clusters literally, it implies that the entire catalog of The Mysterious Package Company may only be 17 to 19 products deep. However, this is purely speculation.

The main outcomes of this of the analysis of The Mysterious Package Company social media footprint is as follows:
- Facebook, product announcements to a large customer base
- Twitter, customer support for audience members who may be confused by the nature of the product
- Instagram, storytelling, and advertising.
